```python
import pandas as pd
import numpy as np
```

# Projeto 1

## Etapa 1

```python
df_vendas = pd.read_json('/content/drive/MyDrive/56 Pandas: transformação e manipulação de dados/dados_vendas_clientes.json')
df_vendas
```

```python
df_vendas = pd.json_normalize(df_vendas['dados_vendas'])
df_vendas
```

## Etapa 2

```python
colunas_vendas = list(df_vendas.columns)
colunas_vendas
```

```python
df_vendas = df_vendas.explode(colunas_vendas[1:])
df_vendas
```

```python
df_vendas.reset_index(inplace=True, drop=True)
df_vendas
```

```python
df_vendas.info()
```

```python
df_vendas['Valor da compra'] = df_vendas['Valor da compra'].apply(
    lambda x: x.replace('R$', '').replace(',', '').strip())
df_vendas['Valor da compra']
```

```python
df_vendas['Valor da compra'] = df_vendas['Valor da compra'].astype(np.float64)
```

```python
df_vendas.info()
```

## Etapa 3

```python
df_vendas
```

```python
df_vendas['Cliente'] = df_vendas['Cliente'].str.lower()
df_vendas['Cliente'] = df_vendas['Cliente'].str.replace('[^a-zA-Z\ã]', ' ', regex=True)
df_vendas['Cliente'] = df_vendas['Cliente'].str.strip()
df_vendas
```

## Etapa 4

```python
df_vendas['Data de venda'] = pd.to_datetime(df_vendas['Data de venda'], format='%d/%m/%Y')
df_vendas
```

```python
total_compras = df_vendas.groupby(['Cliente'])['Valor da compra'].sum()
total_compras
```

# Projeto 2

## Etapa 1

```python
df_locacao = pd.read_json('/content/drive/MyDrive/56 Pandas: transformação e manipulação de dados/dados_locacao_imoveis.json')
df_locacao
```

```python
df_locacao = pd.json_normalize(df_locacao['dados_locacao'])
df_locacao
```

## Etapa 2

```python
colunas_locacao = list(df_locacao.columns)
df_locacao = df_locacao.explode(colunas_locacao[1:])
df_locacao = df_locacao.reset_index(drop=True)
df_locacao
```

```python
df_locacao['valor_aluguel'] = df_locacao['valor_aluguel'].apply(
    lambda x: x.replace('$', '').replace(',', '').replace('reais', '').strip())
df_locacao['valor_aluguel'] = df_locacao['valor_aluguel'].astype(np.float64)
df_locacao.info()
```

## Etapa 3

```python
df_locacao
```

```python
df_locacao['apartamento'] = df_locacao['apartamento'].str.replace('(blocoAP)', '')
df_locacao
```

## Etapa 4

```python
df_locacao.info()
```

```python
df_locacao['datas_combinadas_pagamento'] = pd.to_datetime(df_locacao['datas_combinadas_pagamento'], format='%d/%m/%Y')
df_locacao['datas_de_pagamento'] = pd.to_datetime(df_locacao['datas_de_pagamento'], format='%d/%m/%Y')
df_locacao.info()
```

```python
df_locacao['atraso'] = (
    df_locacao['datas_de_pagamento'] - df_locacao['datas_combinadas_pagamento'])
df_locacao
```

```python
media_atraso = df_locacao.groupby(['apartamento'])['atraso'].mean()
media_atraso
```
